# Exploratory Data Analysis – Credit Card Transactions 

**Objective**: Explore the anonymized bank credit card transaction dataset, quantify the extreme class imbalance, and uncover initial patterns in the PCA-transformed features (`V1`–`V28`), `Time`, and `Amount` that discriminate fraudulent transactions.

This notebook focuses on **data cleaning** and **exploratory data analysis** for the credit card dataset.
______
## 1. Setup and Data Loading


```python
import sys
import os
# Add project root (one directory above "notebooks")
sys.path.append(os.path.abspath(".."))
```

```python
#add necessary imports
import pandas as pd 
import matplotlib.pyplot as plt
import seaborn as sns
from src.data_cleaning import remove_duplicates,remove_missing_values
```

```python
 #load creditcard.csv into notebook
credit_df = pd.read_csv('../data/raw/creditcard.csv') 
# correct data type
credit_df['Class'] = credit_df['Class'].astype(int)
```

```python
# quick check
display(credit_df.head())
display(credit_df.info())
display(credit_df.describe())
```

## 2. Data Cleaning

```python
# drop missing values (Justification: Data integrity is critical for fraud detection)
credit_df = remove_missing_values(credit_df)
```

```python
# Clean Duplicates
credit_df = remove_duplicates(credit_df)
```

# 3. Exploratory Data Analysis

```python
# ============================
# Univariate EDA - Credit Card
# ============================

# Hist for transaction amount
import numpy as np

# Create side-by-side comparison
fig, ax = plt.subplots(1, 2, figsize=(16, 5))
# Plot 1: Original (Skewed)
sns.histplot(credit_df['Amount'], bins=50, kde=True, ax=ax[0])
ax[0].set_title("Original Amount Distribution")
# Plot 2: Log Transformed (Normalized)
# We use np.log1p to avoid errors with 0 values
sns.histplot(np.log1p(credit_df['Amount']), bins=50, kde=True, ax=ax[1])
ax[1].set_title("Log-Transformed Amount Distribution")
ax[1].set_xlabel("Log(Amount + 1)")

plt.show()
# Hist for time
plt.figure(figsize=(8,4))
sns.histplot(credit_df['Time'], bins=50, kde=True)
plt.title("Distribution of Time")
plt.xlabel("Time (seconds)")
plt.ylabel("Count")
plt.show()
```

```python
# ==================================
# Class Imbalance Visualization
# ==================================

sns.countplot(x='Class', data=credit_df)
plt.title("Fraud vs Non-Fraud Count")
plt.xlabel("Class (0 = Legit, 1 = Fraud)")
plt.ylabel("Count")
plt.show()

fraud_rate = credit_df['Class'].value_counts(normalize=True) * 100
print("Class Distribution (%):")
print(fraud_rate)
```

```python
# ==================================
# Bivariate Analysis
# ==================================

# Boxplot amount vs fraud
plt.figure(figsize=(8,4))
sns.boxplot(x='Class', y='Amount', data=credit_df)
plt.title("Amount vs Fraud Class")
plt.xlabel("Class")
plt.ylabel("Amount")
plt.show()


# Scatter for sample PCA features vs class
sns.scatterplot(x='V1', y='V2', hue='Class', data=credit_df.sample(5000))
plt.title("V1 vs V2 by Class (Sampled)")
plt.show()
```
